```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.ticker as mtick  
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

df = pd.read_excel("../data/raw/Telco_customer_churn.xlsx")
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df["Churn Label"].value_counts()
```

```python
df[df["Churn Label"]=="Yes"]["Churn Reason"].isnull().sum()
```

```python
df["Total Charges"] = pd.to_numeric(
    df["Total Charges"],
    errors="coerce"
)
```

```python
df["Total Charges"].isnull().sum()
```

```python
df["Revenue_Per_Month"] = (
    df["Total Charges"] /
    (df["Tenure Months"] + 1)
)
```

```python
df["Tenure_Group"] = pd.cut(
    df["Tenure Months"],
    bins=[0,12,24,48,72],
    labels=[
        "0-1 Year",
        "1-2 Years",
        "2-4 Years",
        "4-6 Years"
    ]
)
```

```python
df["Charge_Group"] = pd.qcut(
    df["Monthly Charges"],
    q=4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High"
    ]
)
```

```python
cols_to_drop = [
    "Count",
    "Country",
    "State",
    "Lat Long",
    "Latitude",
    "Longitude",
    "Zip Code"
]

df_clean = df.drop(columns=cols_to_drop)
```
